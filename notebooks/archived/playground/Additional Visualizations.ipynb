```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

sns.set('paper', style='whitegrid')
```

```python
X = np.random.random(size=(7, 7))
X = np.triu(X)
X[np.tril_indices(7)] = np.nan
sns.heatmap(X, square=True, cmap='Blues', robust=True)
```

```python

X = pd.DataFrame(np.random.random((7, 125))).reset_index()
# sns.lineplot()
X = X.melt(id_vars='index')
X.columns = ['network', 'time', 'signal']

X['signal'] = X['signal'] + X['network'] / 1.4

sns.set('paper', style='white')

sns.lineplot(data=X, x='time', y='signal', hue='network', legend=False, lw=2)
```
